# Using the WeatherLink API for Download

This notebook sets up the WeatherLink API v2 apisignature for download of the Sunapee weather stations.

```python
import collections
import hashlib
import hmac
import time
import pip
import requests
import json
import pandas
```

First, let's look at the Weather Station details. Parameters that are needed are user key, secret and current time. First, set the parameters for this request. Parameters for each request are detailed here: https://weatherlink.github.io/v2-api/api-reference

```python
parametersStation = {
  "api-key": "fhbciecrnnzumskmacfkvvmnbjpzuki6", #this is my key
  "api-secret": "ur8emqsds5jyshegr3okeniw8rp6gfgn", #and my 'secret' - these need to be updated for other purposes
  "t": int(time.time())
}
```

```python
parametersStation = collections.OrderedDict(sorted(parametersStation.items()))

for key in parametersStation:
  print("Parameter name: \"{}\" has value \"{}\"".format(key, parametersStation[key]))

apiSecret = parametersStation["api-secret"];
parametersStation.pop("api-secret", None)

data = ""
for key in parametersStation:
  data = data + key + str(parametersStation[key])

apiSignature = hmac.new(
  apiSecret.encode('utf-8'),
  data.encode('utf-8'),
  hashlib.sha256
).hexdigest()

getAddressStations = "https://api.weatherlink.com/v2/stations/?api-key={}&api-signature={}&t={}".format(parametersStation["api-key"], apiSignature, parametersStation["t"])
```

```python
station_list = requests.get(getAddressStations)
print(station_list.json())
```

And then let's get a single day of historic data from Herrick Cove (still need to deal with timezone issues):

```python
parametersHistoricHC = {
  "api-key": "fhbciecrnnzumskmacfkvvmnbjpzuki6", #unique user key
  "api-secret": "ur8emqsds5jyshegr3okeniw8rp6gfgn", #user pass
  "station-id": "73101",  #station id from stations list
  "t": int(time.time()), #current time
  "start-timestamp": int(time.mktime(time.strptime("2020-10-05 00:00 -0400", "%Y-%m-%d %H:%M %z"))), #start of timestamp
  "end-timestamp": int(time.mktime(time.strptime("2020-10-06 00:00 -0400", "%Y-%m-%d %H:%M %z"))) #end of timestamp
}
```

```python
parametersHistoricHC = collections.OrderedDict(sorted(parametersHistoricHC.items()))

for key in parametersHistoricHC:
  print("Parameter name: \"{}\" has value \"{}\"".format(key, parametersHistoricHC[key]))

apiSecret = parametersHistoricHC["api-secret"];
parametersHistoricHC.pop("api-secret", None)

data = ""
for key in parametersHistoricHC:
  data = data + key + str(parametersHistoricHC[key])

apiSignature = hmac.new(
  apiSecret.encode('utf-8'),
  data.encode('utf-8'),
  hashlib.sha256
).hexdigest()

getAddressHistoricHC = "https://api.weatherlink.com/v2/historic/{}?api-key={}&api-signature={}&t={}&start-timestamp={}&end-timestamp={}".format(parametersHistoricHC["station-id"], parametersHistoricHC["api-key"], apiSignature, parametersHistoricHC["t"], parametersHistoricHC["start-timestamp"], parametersHistoricHC["end-timestamp"])
```

```python
historic_HC = requests.get(getAddressHistoricHC)
print(historic_HC.json())
```

Save the json file

```python
with open("api-output/historic_HC.json", "w") as write_file:
    json.dump(historic_HC.json(), write_file)
```

Now read in and deserialize:

```python
with open("api-output/historic_HC.json", "r") as read_file:
    historic_HC_readable = json.load(read_file)

type(historic_HC_readable) #this should be 'dict' - if not, you did it wrong.
```

Let's look at the data in a more readable way:

```python
print(json.dumps(historic_HC_readable, indent = 4, sort_keys=True))
```

And now to save that as a flat file using pandas:

```python
hc_df = pandas.json_normalize(historic_HC_readable["sensors"][0]["data"], max_level=0)
print(hc_df)
```

Reformat time in new column.

```python
hc_df["utc_datetime"] = pandas.to_datetime(hc_df.ts, unit = 's', origin='unix')
print(hc_df)
```

```python
with pandas.ExcelWriter('api-output/hc_historic_trial.xlsx') as writer:
    hc_df.to_excel(writer)
```
